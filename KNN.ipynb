```pyspark
from pyspark.sql import SparkSession

# Initialize a Spark session (if not already done)
spark = SparkSession.builder \
    .appName("ModelTraining") \
    .getOrCreate()

# Define your S3 URI (replace with your bucket and path)
s3_path = "s3://aws-emr-studio-201107706822-us-east-1/1746053900403/e-6HYCE3WANA8IROUS7NBVDSOU4/processed/final_windows/final_6"

# Read the Parquet file from S3 into a PySpark DataFrame
user_df = spark.read.parquet(s3_path)

# Show the top rows to confirm the data is loaded
user_df.show()
```

```pyspark
df_reduced = user_df.select("UserID", "job_vector")
df_reduced.show()
```

```pyspark
jobs_df = spark.read.parquet("s3://aws-emr-studio-201107706822-us-east-1/1746053900403/e-6HYCE3WANA8IROUS7NBVDSOU4/processed/windows/window=6/")
jobs_df.show()
```

```pyspark
applied_df = spark.read.option("header", True).option("sep", "\t").csv("s3://aws-emr-studio-201107706822-us-east-1/1746053900403/e-6HYCE3WANA8IROUS7NBVDSOU4/apps.tsv")
applied_df.show()
```

```pyspark
from pyspark.ml.linalg import Vectors, DenseVector
from pyspark.sql.functions import udf, col, avg
from pyspark.sql.types import ArrayType, DoubleType
```

```pyspark
# Convert DenseVector to array for averaging
vector_to_array_udf = udf(lambda v: v.toArray().tolist(), ArrayType(DoubleType()))
df_array = df_reduced.withColumn("job_vector_array", vector_to_array_udf("job_vector"))
```

```pyspark
df_array.show()
```

```pyspark
# Compute average for each vector dimension across all jobs per user
avg_exprs = [avg(col("job_vector_array")[i]).alias(f"avg_{i}") for i in range(100)]
user_avg_vectors = df_array.groupBy("UserID").agg(*avg_exprs)
```

```pyspark
from pyspark.ml.linalg import Vectors, VectorUDT
```

```pyspark
def to_dense_vector(*vals):
    return Vectors.dense(vals)

to_vector_udf = udf(to_dense_vector, VectorUDT())

user_avg_vectors = user_avg_vectors.withColumn(
    "user_vector_avg", to_vector_udf(*[col(f"avg_{i}") for i in range(100)])
).select("UserID", "user_vector_avg")
```

```pyspark
user_avg_vectors.show()
```

```pyspark
from pyspark.ml.feature import BucketedRandomProjectionLSH
```

```pyspark
lsh = BucketedRandomProjectionLSH(
    inputCol="job_vector", 
    outputCol="hashes", 
    bucketLength=2.0,
    numHashTables=3
)

# Fit model on the job vectors
job_model = lsh.fit(jobs_df)  
```

```pyspark
test_users = applied_df.filter(col("WindowID") == 6).select("UserID").distinct()
test_user_embeds = user_avg_vectors.join(test_users, on="UserID")
```

```pyspark
test_user_embeds.show()
```

```pyspark
batch_size = 1000
total = test_user_embeds.count()
recommendations = []
```

```pyspark
import pyspark.sql.functions as F
```

```pyspark
limited_users_df = test_user_embeds.limit(100)
```

```pyspark
user_ids = [row["UserID"] for row in limited_users_df.select("UserID").collect()]
k = 5
all_recommendations = []

for user_id in user_ids:
    # Filter Spark DF to get one user's embedding
    user_row = test_user_embeds.filter(F.col("UserID") == user_id).first()
    
    if user_row is None:
        continue
    
    user_vec = user_row["user_vector_avg"]

    # Find nearest jobs using the vector
    neighbors = job_model.approxNearestNeighbors(jobs_df, user_vec, k)
    neighbors = neighbors.withColumn("UserID", F.lit(user_id))
    selected = neighbors.select("UserID", "JobID", "distCol").collect()
    
    all_recommendations.extend(selected)
```

```pyspark
all_recommendations
```

```pyspark
recommendation_df = spark.createDataFrame(all_recommendations)
recommendation_df.show(5)
```

```pyspark
actual_apps = applied_df.filter(col("WindowID") == 6).select("UserID", "JobID").distinct()
```

```pyspark
# Join recommended and actual
joined = recommendation_df.join(actual_apps, on=["UserID", "JobID"], how="left_outer").withColumn("label", F.when(F.col("JobID").isNotNull(), 1).otherwise(0))

# Precision@k per user
precision_df = joined.groupBy("UserID").agg(F.sum("label").alias("correct"), F.count("JobID").alias("recommended")).withColumn("precision_at_k", F.col("correct") / F.col("recommended"))

# Average Precision@k
avg_precision = precision_df.select(F.avg("precision_at_k")).first()[0]
```
